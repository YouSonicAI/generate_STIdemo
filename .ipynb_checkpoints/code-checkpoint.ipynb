```python
import cv2
videos_path = "./渐进式.mp4"
vidcap = cv2.VideoCapture(videos_path)
success, image = vidcap.read()
time_interval = 1 
frames_save_path = r"C:\Users\17579\Desktop\Yousonic\STI_Video_Demo\Self\Image\down"
success, image = vidcap.read()
```

```python
count = 0
while success:
    success, image = vidcap.read()
    count += 1
    start = time.time()
    if count % time_interval == 0:
      cv2.imencode('.jpg', image)[1].tofile(frames_save_path + "/frame%d.jpg" % count)
```
